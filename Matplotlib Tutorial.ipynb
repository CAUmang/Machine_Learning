```python
#Compute the x and y cordinates for points on a sine curve

x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)

plt.title("sin wave form")

#plot the points using matplotlib
plt.plot(x,y)
```

```python
#subplot
#Compute the x and y coordinates for points on sine and cosine curves

x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#Setup a subplot grid that has height 2 and width 1
#and set the first such subplot as active.

plt.subplot(2,1,1)

#Make the first plot
plt.plot(x,y_sin, 'r--')
plt.title('sine')

#Set the second plot as active, and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos, 'g--')
plt.title('cosine')
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')
```

```python
#Box Plot in Matplotlib

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```
